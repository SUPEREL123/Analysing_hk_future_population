# Analysing the prediction of Hong Kong Population from 2021 to 2031

## data from https://data.gov.hk/en-data/dataset/hk-pland-pland1-projections-of-population-distribution-2023-to-2031/resource/38a116d8-5053-4243-bd9d-188a7248c79b

## load the data and also the functions

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("Tables_21_based_Table_1.csv")
df.drop('區議會分區／主要區域', inplace=True, axis=1)
```

## Draw a pie chart showing the comparison of different districts in same year

### There are not much change between the comparison of different districts in hong kong during 10 years

```python
def year_pop_pie(year):
  year_index = year - 2020
  column_name = list(df.iloc[:len(df)-3,0].values)
  pop = list(df.iloc[:len(df)-3,year_index].values)
  plt.pie(pop, labels = column_name)
  plt.title(f"pie chart for {year}")
  plt.show()

year_pop_pie(2021)
year_pop_pie(2030)
```

## plot line graphs about the change of population in different districts with increasing the year

### Sham Shui Po has interesting data. It shows a bug increase from 2021 to 2024 and start decrease in 2024. Also the population in Sham Shui Po increased in 2026 again.



```python
column_names = list(df.iloc[:,0].values)
for i in range (0,len(column_names)):
  df.iloc[i,1:].plot()
  plt.title(df.iloc[i,0])
  plt.show()
```

## drawing a bar chart showing the percentage change of population from 2021 to 2031

### The population in Islands increase a lot. Probably it is because of the "Lantau Tomorrow Vision"

```python
first_year = df.iloc[:,1].values
last_year = df.iloc[:,11].values

percentage_change_of_each_year = 100*(last_year - first_year)/first_year
percentage_change_of_each_year

district = list(df.iloc[:,0])

plt.bar(district, percentage_change_of_each_year)
plt.xticks(rotation=45, ha="right")
plt.xlabel("District")
plt.ylabel("Percentage change of population")
plt.title("percentage change in different district over 10 years")

plt.show()
```
